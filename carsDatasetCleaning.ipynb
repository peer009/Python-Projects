1. Find all Null Values in the dataset. If there is any null value in any column, then fill it with the mean of that column

```python
#Find all Null values in any column

cars.isnull().sum()
```

```python
#Fill Null values with the mean of the column

cars["Cylinders"].fillna(cars["Cylinders"].mean(), inplace = True)
cars["Weight"].fillna(cars["Weight"].mean(), inplace = True)
```

```python
#No more Null values

cars.isnull().sum()
```

2. Check what are the different types of 'Make' in the dataset and what is the count(occurrence) of each 'Make' in the data 

```python
cars["Make"].value_counts()
```

3. Show all the records where 'Origin is Asia or Europe

4. Remove all the records with 'Weight' above 4000

```python
cars.head()
```

```python
#Showing all cars with Weight above 4000

cars[cars["Weight"] > 4000]
```

```python
#Removing all cars with a weight above 4000

cars[~(cars["Weight"] > 4000)]
```

5. Increase all the values of 'MPG_City' column by 3

```python
#Using lambda function to increase column values by 3

cars["MPG_City"] = cars["MPG_City"].apply(lambda x:x+3)
```
