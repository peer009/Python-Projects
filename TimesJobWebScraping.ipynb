```python
from bs4 import BeautifulSoup
import requests
import csv

# Connecting to website for scraping
url = "https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&txtKeywords=python&txtLocation="

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36",
    "Accept-Encoding": "gzip, deflate", "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8",
    "DNT": "1", "Connection": "close", "Upgrade-Insecure-Requests": "1"}

page = requests.get(url, headers=headers)
soup = BeautifulSoup(page.content, "html.parser")

print(soup)
```

```python
#Fetching li tags which contain the main information we want to scrape 

jobs = soup.find_all("li", class_="clearfix job-bx wht-shd-bx")
```

```python
# Create Excel file to save data scraped(run this once)

header = ["Job Title", "Company Name", "Skills", "Experience", "Job Page"]


with open("pythonjobsavaliable.csv", "w", newline="", encoding="UTF8") as f:
    writer = csv.writer(f)
    writer.writerow(header)
    
```

```python
# Looping through html to find specific tags and store their values in the excel file created

for job in jobs:
    job_title = job.find("h2", class_="").text
    company_name = job.find("h3", class_="joblist-comp-name").text.replace(" ", "")
    skills = job.find("span", "srp-skills").text.replace(" ", "")
    experience = job.find("li", class_="").text.replace(" ", "").strip()[11:]
    job_page = job.header.h2.a["href"]


    header = ["Job Title", "Company Name", "Skills", "Experience", "Job Page"]
    data = [job_title.strip(), company_name.strip(), skills.strip(), experience, job_page]

    # Add new data to the Excel file created
    with open("pythonjobsavaliable.csv", "a+", newline="", encoding="UTF8") as f:
        writer = csv.writer(f)
        writer.writerow(data)
```
